```python
#The maximum Sales amd Profit generated by Standard Class
#We need to focus on Same Day delivery in order to increase profit
```

```python
#Analyzing Sales and profit by different Regions
data.groupby('Region')['Sales','Profit'].agg(sum).plot(kind = 'bar',color=['brown','green'])
plt.title('Total Sales and Profits by different Regions')
plt.ylabel('Total Sales and Profits')
plt.show()
```

```python
#Analyzing Sales and Profit by different categories
data.groupby('Category')['Sales','Profit'].agg(sum).plot(kind = 'bar',color=['teal','purple'])
plt.title('Total Sales and Profits by different Categories')
plt.ylabel('Total Sales and Profits')
plt.show()
```

```python
#Analyzing Sales and Profit by different Sub-Categories
data.groupby('Sub_Category')['Sales','Profit'].agg(sum).plot(kind = 'barh',color=['coral','magenta'])
plt.title('Total Sales and Profits by different Sub-Categories')
plt.ylabel('Total Sales and Profits')
plt.show()
```
